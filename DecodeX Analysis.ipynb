```python
import pandas as pd

# File paths
known_route_path = "/content/drive/MyDrive/Colab Notebooks/Known route.csv"
ongoing_data_path = "/content/drive/MyDrive/Colab Notebooks/On going data.csv"

# Load datasets
known_route_df = pd.read_csv(known_route_path)
ongoing_data_df = pd.read_csv(ongoing_data_path)

# Display info and first few rows with proper formatting
print("=== Known Route DataFrame Info ===")
known_route_df.info()
print("\n=== First 5 Rows of Known Route DataFrame ===\n", known_route_df.head())

print("\n\n=== Ongoing Data DataFrame Info ===")
ongoing_data_df.info()
print("\n=== First 5 Rows of Ongoing Data DataFrame ===\n", ongoing_data_df.head())
```

```python
# Extract first and last location of each unique Route_ID
known_locations_df = known_route_df.groupby("Route_ID").agg(
    first_latitude=("latitude", "first"),
    first_longitude=("longitude", "first"),
    last_latitude=("latitude", "last"),
    last_longitude=("longitude", "last")
).reset_index()

# Display extracted known locations
known_locations_df
```

```python
from geopy.distance import geodesic

# Function to calculate distance
def calculate_distance(row):
    start = (row["first_latitude"], row["first_longitude"])
    end = (row["last_latitude"], row["last_longitude"])
    return geodesic(start, end).kilometers

# Apply function to calculate distances
known_locations_df["distance_km"] = known_locations_df.apply(calculate_distance, axis=1)

# Display result
print(known_locations_df)
```

```python
route_id_counts = known_locations_df['Route_ID'].value_counts()
print(route_id_counts)
```

```python
import pandas as pd

# Load the known route and ongoing data files
known_route_path = "/content/drive/MyDrive/Colab Notebooks/Known route.csv"
ongoing_data_path = "/content/drive/MyDrive/Colab Notebooks/On going data.csv"

known_route_df = pd.read_csv(known_route_path)
ongoing_data_df = pd.read_csv(ongoing_data_path)

# Display the first few rows of each dataset to understand the structure
known_route_df.head(), ongoing_data_df.head()
```

```python
from scipy.spatial import cKDTree
import numpy as np

# Extract coordinates from known route
known_coords = known_route_df[["latitude", "longitude"]].values
known_tree = cKDTree(known_coords)  # Create a spatial tree for efficient lookup

# Extract coordinates from ongoing data
ongoing_coords = ongoing_data_df[["latitude", "longitude"]].values

# Define a threshold distance (in degrees, rough approximation for small distances)
threshold = 0.1  # Adjust based on precision needed (~10km)

# Find the nearest known route point for each ongoing pallet location
distances, _ = known_tree.query(ongoing_coords)

# Flag anomalies (pallets outside the known route threshold)
ongoing_data_df["Off_Route"] = distances > threshold

# Extract off-route pallets
off_route_pallets = ongoing_data_df[ongoing_data_df["Off_Route"]]

# Display a summary
off_route_pallets
```

```python
import folium
from scipy.spatial import cKDTree
import numpy as np

# Extract coordinates from known route
known_coords = known_route_df[["latitude", "longitude"]].values
known_tree = cKDTree(known_coords)  # Create a spatial tree for efficient lookup

# Extract coordinates from ongoing data
ongoing_coords = ongoing_data_df[["latitude", "longitude"]].values

# Define a threshold distance (~5 km in degrees)
threshold = 0.1  # Approx. 10 km

# Find the nearest known route point for each ongoing pallet location
distances, _ = known_tree.query(ongoing_coords)

# Flag anomalies (pallets outside the known route threshold)
ongoing_data_df["Off_Route"] = distances > threshold

# Separate off-route and on-route pallets
off_route_pallets = ongoing_data_df[ongoing_data_df["Off_Route"]]
on_route_pallets = ongoing_data_df[~ongoing_data_df["Off_Route"]]

# Generate a map centered at the mean location
map_center = [ongoing_data_df["latitude"].mean(), ongoing_data_df["longitude"].mean()]
m = folium.Map(location=map_center, zoom_start=6)

# Plot known routes as a blue line
route_coords = list(zip(known_route_df["latitude"], known_route_df["longitude"]))
folium.PolyLine(route_coords, color="blue", weight=2, opacity=0.7, tooltip="Known Route").add_to(m)

# Plot on-route pallets as small green circles
for _, row in on_route_pallets.iterrows():
    folium.CircleMarker(
        location=[row["latitude"], row["longitude"]],
        radius=2,  # Smaller size
        color="green",
        fill=True,
        fill_opacity=0.7
    ).add_to(m)

# Plot off-route pallets as small red circles
for _, row in off_route_pallets.iterrows():
    folium.CircleMarker(
        location=[row["latitude"], row["longitude"]],
        radius=2,  # Smaller size
        color="red",
        fill=True,
        fill_opacity=0.7
    ).add_to(m)

# Display map
m
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Extract relevant features for training
features_known = known_route_df[["latitude", "longitude"]].copy()

# Standardize the data for better training performance
scaler = StandardScaler()
features_known_scaled = scaler.fit_transform(features_known)

# Display sample data
print("ðŸ“Š Sample of Training Data (Known Routes):")
print(pd.DataFrame(features_known_scaled, columns=["latitude", "longitude"]).head())
```

```python
from sklearn.ensemble import IsolationForest

# Train Isolation Forest model
iso_forest = IsolationForest(contamination=0.05, random_state=42)
iso_forest.fit(features_known_scaled)

print("âœ… Model trained on known routes!")
```

```python
# Convert 'Date' and 'Time' into proper datetime format if needed
if "datetime" not in ongoing_data_df.columns:
    if "Date" in ongoing_data_df.columns and "Time" in ongoing_data_df.columns:
        ongoing_data_df["datetime"] = pd.to_datetime(ongoing_data_df["Date"] + " " + ongoing_data_df["Time"], errors="coerce")
        ongoing_data_df.drop(columns=["Date", "Time"], inplace=True)  # Drop old columns after conversion
    elif "timestamp" in ongoing_data_df.columns:  # If timestamp column exists, rename it
        ongoing_data_df.rename(columns={"timestamp": "datetime"}, inplace=True)
    else:
        raise KeyError("âŒ 'datetime' column is missing and cannot be created! Check your dataset.")
```

```python
# Extract relevant features from ongoing data
features_ongoing = ongoing_data_df[["latitude", "longitude"]].copy()

# Standardize using the same scaler from training
features_ongoing_scaled = scaler.transform(features_ongoing)

# Predict anomalies (-1 = anomaly, 1 = normal)
ongoing_data_df["Anomaly_Score"] = iso_forest.predict(features_ongoing_scaled)
ongoing_data_df["Anomaly"] = ongoing_data_df["Anomaly_Score"].apply(lambda x: "Anomaly" if x == -1 else "Normal")

# Display detected anomalies
anomalies = ongoing_data_df[ongoing_data_df["Anomaly"] == "Anomaly"]
print("ðŸš¨ Detected Anomalous Pallets:")
print(anomalies[["Pallet_ID", "datetime", "latitude", "longitude", "Anomaly"]])
```

```python
import folium
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

# Extract features for anomaly detection
features_ongoing = ongoing_data_df[["latitude", "longitude"]].copy()

# Standardize the data
scaler = StandardScaler()
features_ongoing_scaled = scaler.fit_transform(features_ongoing)

# Train Isolation Forest model
iso_forest = IsolationForest(n_estimators=200, contamination=0.02, random_state=42)
ongoing_data_df["Anomaly_Score"] = iso_forest.fit_predict(features_ongoing_scaled)

# Label anomalies (-1 = anomaly, 1 = normal)
ongoing_data_df["Anomaly"] = ongoing_data_df["Anomaly_Score"].apply(lambda x: "Anomaly" if x == -1 else "Normal")

# Separate normal and anomalous pallets
anomalous_pallets = ongoing_data_df[ongoing_data_df["Anomaly"] == "Anomaly"]
normal_pallets = ongoing_data_df[ongoing_data_df["Anomaly"] == "Normal"]

# Create a map centered at the mean location of ongoing pallets
map_center = [ongoing_data_df["latitude"].mean(), ongoing_data_df["longitude"].mean()]
m = folium.Map(location=map_center, zoom_start=6)

# Plot known route as a blue line
route_coords = list(zip(known_route_df["latitude"], known_route_df["longitude"]))
folium.PolyLine(route_coords, color="blue", weight=2, opacity=0.7, tooltip="Known Route").add_to(m)

# Plot normal pallets as small green circles
for _, row in normal_pallets.iterrows():
    folium.CircleMarker(
        location=[row["latitude"], row["longitude"]],
        radius=2,  # Small size
        color="green",
        fill=True,
        fill_opacity=0.7
    ).add_to(m)

# Plot anomalous pallets as small red circles
for _, row in anomalous_pallets.iterrows():
    folium.CircleMarker(
        location=[row["latitude"], row["longitude"]],
        radius=2,  # Small size
        color="red",
        fill=True,
        fill_opacity=0.7
    ).add_to(m)

# Display the map
m
```
